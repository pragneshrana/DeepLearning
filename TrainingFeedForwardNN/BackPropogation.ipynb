# Chain rule of derivative 

-  $x -(w_1)-> sig  -> h1 -(w_2) -> sig -> h2 -(w_3)-> sig -> \hat y --> CE -> -log \hat y L$

Means $L = f(x,w_1,w_2,w_3)$

$\frac{\partial L}{\partial w_2} = \frac{\partial L}{\partial \hat y} \frac{\partial \hat y}{\partial h_2} \frac{\partial h_2}{\partial w_2} \frac{\partial h_2}{\partial w_2}$

- Computation of the gradients for all the the layers and the inputs are tedius process.
- Each weight of input influences the many neurons and ultimately output.

Now, for square loss, $L = (y_1 - \hat y_1)^2 + (y_2 - \hat y_2)^2$

As, we want to update the weight $W_{212}$, we need to find out the derivative of loss function w.r.t that weight.

so,
$\frac{\partial L}{\partial w_{212}} = \frac{\partial L}{\partial a_{21}} \frac{\partial a_{21}}{\partial w_{212}} =\frac{\partial L}{\partial \hat y_1} \frac{\partial \hat y_1}{\partial a_{21}} \frac{\partial a_{21}}{\partial w_{212}}$

$\frac{\partial L}{\partial \hat y_1} = -2(y_1 - \hat y_1)$

**Softmax** is applied to the final layer so,
$\hat y_1 = \frac{\exp^{a_{21}}}{\exp^{a_{21}}+\exp^{a_{22}}} = \frac{1}{1+\exp^{-a_{21}+a_{22}}}$

now,
$\frac{\partial \hat y_1}{\partial a_{21}} = \bigg(\frac{1}{1+\exp^{-a_{21}+a_{22}}}\bigg)^2 \cdot \exp^{-a_{21}+a_{22}}\cdot (- 1) = \frac{1}{1+\exp^{-a_{21}+a_{22}}} \cdot \frac{\exp^{-a_{21}+a_{22}}}{1+\exp^{-a_{21}+a_{22}}}\cdot (-1) = \hat y_1 \cdot(1-\hat y_1) \cdot (-1)$

now,
$a_{21} = w_{211} h_{11} + w_{212} h_{12} + w_{213} h_{13} + w_{214} h_{14}$

$\frac{\partial a_{21}}{\partial w_{212}} = h_{12}$

$\frac{\partial L}{\partial w_{212}} =\frac{\partial L}{\partial \hat y_1} \frac{\partial \hat y_1}{\partial a_{21}} \frac{\partial a_{21}}{\partial w_{212}} = -2\cdot (y_1 - \hat y_1) \cdot \hat y_1 \cdot(1-\hat y_1) \cdot (- 1) \cdot 
h_{12}$

$\frac{\partial \hat y_1}{\partial a_{21}} = -2\cdot (y_1 - \hat y_1) \cdot \hat y_1 \cdot(1-\hat y_1) \cdot (- 1) \cdot h_{12}$

$\frac{\partial \hat y_1}{\partial a_{21}} = -2\cdot (1 - 0.77) \cdot 0.77 \cdot(1-0.77) \cdot (-1) \cdot 0.80$




$a_1 = W_1 \cdot x + b_1 = [2.9 \space 1.4 \space 2.1 \space 2.3]$

$h_1 = sigmoid(a_1) = [0.95 \space 0.80 \space 0.89 \space 0.91]$

$a_2 = W_2 \cdot h1+ b_2 = [1.66 \space 0.45 ]$

$h_1 = sigmoid(a_2) = [0.77 \space 0.23 ]$


$\frac{\partial \hat y_1}{\partial a_{21}} = -2\cdot (y_1 - \hat y_1) \cdot \hat y_1 \cdot(1-\hat y_1) \cdot (- 1) \cdot h_{12}$

$\frac{\partial \hat y_1}{\partial a_{21}} = -2\cdot (1 - 0.77) \cdot 0.77 \cdot(1-0.77) \cdot (-1) \cdot 0.80 = -0.46 \cdot -0.079695 \cdot 0.8 = -0.0651$

we can update by,

$W_{212} = W_{212} - \eta (-0.0651) $
