# What is the way to come up with complex function?


- Combination of fundamental function can generate complex funxtion.
- One of the basic block is sigmoid function.
    - $f(x,w) = \frac{1}{1+\exp(wx+b)}$

- Combining many such function can actually generate complex function.

- Method is generate such complex function by fundamental block leads to deep neural network.

-

# Unniversal approximation theorem.

- It says that using DNN any complex function can be obtained.

- Combinaion of simple function (same family) can generate any complex function.

- $f_1(x_a,w_a) = \frac{1}{1+\exp(w_ax_a+c)}$

- $f_2(x_b,w_b) = \frac{1}{1+\exp(w_bx_b+d)}$

- Combining(linear combination) of such sigmoid function can generate more compled function which has different nature than simple sigmoid function. 

- $f(x_ab,w_ab) = \frac{1}{1+\exp(w_ax_a+c)} \pm \frac{1}{1+\exp(w_bx_b+d)}$ ---  which is different than sigmoid function.

# Overall Neural Net and Output function

- Model with two input layer can be expressed as,

$\hat y = \mathcal O(W_3\underbrace{g\underbrace{(W_2\underbrace{g(\underbrace{W_1 X + b_1)}_\text{$a_1$}}_\text{$h_1$}+b_2)}_\text{$a_2$}}_\text{$h_2$}+b_3)$

- Output layer activation function can be decided the task. One of the function is softmax function.

- $\mathcal O = \frac{\exp^{x_i}}{\sum_j \exp^{x_j}}$  whereas, i denotes one of the output.

- To evaluate performance of the network cross entropy loss can be used.
- $\mathcal L (\Phi) = - \sum_i^k y_i \log(\hat y_i) $
