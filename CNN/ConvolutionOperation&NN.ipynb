Convolution in 2-D:

- Convolution operation boils down to taking a given input and re-estimating it as a weighted average of all the inputs around it.

- In 2D consider neighbout along the rows and columns around it.

$s_{ij} = (I*K)_{ij} = \sum_{a=0}^{m-1} \sum_{b=0}^{n-1} I_{i+1,j+b}*K_{a,b}$

where, K refers t the kernal.


$\underbrace{\begin{bmatrix}
a & b & c & d \\
e & f & g & h \\
i & j & k & l \\
\end{bmatrix}}_{Input} * \underbrace{\begin{bmatrix}
w & x \\
y & z \\
\end{bmatrix}}_{Kernel} = \underbrace{\begin{bmatrix}
aw + bx +ey +fz & bw + cx + fy +gz & cw + dx +gy +hz \\
ew+fx+iy+jz & fw+gx+jy+kz & gw+hx+ky+lz \\
\end{bmatrix}}_{Output} $

Terminology :
  - Input : 
    - Width $W_1$
    - Height $H_1$
    - Depth $D_1$
  - Output :
   - Weight $W_0$
   - Height $H_0$
   - Depth $D_0$
  
  - Spatial extent of a filter (F)
  - Fiter Depth is always same as input depth
    - Output depth will be 2D
  - Number of filters - K
  - Padding - P
  - Stride - S

- If kernel is not placed at corner or edges, Relation :
  - $W_0= W_1 - F + 1$
  - $H = H_1 - F + 1$

  - $W_0= \frac{W_1 - F +2P }{S}+1$
