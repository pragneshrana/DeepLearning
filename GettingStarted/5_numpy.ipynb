```
X = np.zeros((4,5))
print(X)
```

```
X = np.eye(4,5) #Identity matrix
print(X)
```

```
X = np.random.random((4,5)) #random matrix
print(X)
```

```
Z = X.T #transpose
print(Z)
```

```
Y = Z.reshape(X.shape[0]*X.shape[1],1)
print(Y)
```
