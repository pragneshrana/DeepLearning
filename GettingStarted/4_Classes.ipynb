#Classes

```
class MobilePhone:
  '''
  this is a sample class to explain how python classes work
  '''

  def __init__(self,name,is_android=False,screen_size=8):
    self.name = name
    self.is_android = is_android
    self.screen_size = screen_size
    self.rating = -1
  def has_rating(self):
    return self.rating > -1
```

```
new_phone = MobilePhone('iphone 12')
```

```
type(new_phone)
```

```
print(new_phone.name, new_phone.is_android)
```

```
new_phone.screen_size = 7
```

```
new_phone.has_rating()
```

```
new_phone.rating = 4
```

#Inheritane

```
class iphone(MobilePhone):
  def __init__(self,name):
    MobilePhone.__init__(self,name,False,4)

  def __str__(self):
    return str(self.name) + " " + str(self.is_android) + " " + str(self.screen_size)
```

```
new_iphone = iphone('iphone 6s')
```

```
new_iphone.is_android
```
