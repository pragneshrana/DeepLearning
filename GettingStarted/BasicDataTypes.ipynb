# Why data type of tuple?
Observe the time difference

```
my_list = list(range(100000000))
tic_list = time.time()
my_list_squared = map(lambda x:x**2,my_list)
toc_list = time.time()
print(toc_list - tic_list)
```

```
my_tuple = tuple(range(100000000))
tic_tuple = time.time()
my_tuple_squared = map(lambda x:x**2,my_tuple)
toc_tuple = time.time()
print(toc_tuple - tic_tuple)
```

# Why set?
Look at the time.

Advantage is to easy to search queries. 

```
import time
my_set = set(range(1000000))
tic_set = time.time()
print(94283 in my_set)
print(100011 in my_set)
print(1111111111 in my_set)
print(11111 in my_set)
toc_set = time.time()
print('Elaspsed time :',toc_set - tic_set)
```

```
my_list = list(range(100000000))
tic_list = time.time()
print(94283 in my_list)
print(100011 in my_list)
print(1111111111 in my_list)
print(11111 in my_list)
toc_list = time.time()
print('Elaspsed time :',toc_list - tic_list)
```
