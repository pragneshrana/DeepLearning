```
def sum_squares(end):
  sum  = 0
  for i in range(end):
    sum += end**2
  return sum

```

```
sum_squares(5)
```

```
def print_sum_square(stop,start=0):
  '''
  This method will return sum of suqare of number between start and stop(excluding) condition
  '''
  sum  = 0
  for i in range(start,stop):
    sum += i**2
  print(sum)
  return sum
```
