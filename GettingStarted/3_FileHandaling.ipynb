```
file  = open('happiness.csv','r') #read mode
```

```
s = file.readline()
```

```
file.close()
```

```
file.readline() #you can't read now
```

```
with open('happiness.csv','r') as file:
  print(file.readlines())
#autommatically close the file
```

```
with open('happiness.csv','r') as file:
  for line in file:
    for word in line.split(','):
      print(word)
```

```
with open('WriteFile.txt','w') as file:
  file.write('hello!')
```
