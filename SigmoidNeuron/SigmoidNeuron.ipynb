# Signoid Family:
 
 Such function does not have abrupt change in function values.  \
 They follow *S* shape. one such example is logistic function. \
$y = \frac{1}{1+\exp^{-(wx+b)}}$

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10,10,0.1)
w = np.ones(len(x)) * 2
b = np.ones(len(x)) * 0.5

one = np.ones(len(x))  #wx+b
y = one/(one+np.exp(-w*x+b)) 
plt.plot(x,y)
plt.xlabel('wx+b')
plt.ylabel('y')
plt.title('Logistic Function')
```

# Loss Function:

- Squared error loss function : $\sum_i (y - \hat y)^2$ \
     This indicates that the data point which has more error need more correction and contribute more to the error. Another such function is cross entropy function.

# Learning Algo:

- Initialize: $w $ & $b$
- Iterate over data: 
    - compute $\hat y$
    - compute Loss $\mathcal{L} (w,b)$
    - $w = w + \Delta w $ \
    - $b = b + \Delta b $ \
    - till satisfied 
  
Now, problem is we need proper guess for $\Delta w $ and $\Delta b $ such that it take smaller steps so that the variation in loss function is chnaged in control manner (with less fluctuation). 

One of the way is to multiply with smaller value $<1$, say 'learning rate'.

- $w = w + \eta \cdot \Delta w $ \
- $b = b + \eta \cdot \Delta b $ \

- Still, the goal is to find $\Delta \theta$ ($\theta$ is general way to represent w and b) such away that loss function should be minimized in next step.

    mathematically, \
    $w \rightarrow w + \eta \cdot \Delta w$ \
    S.T $\mathcal{L} (w) > \mathcal{L} ( w + \eta \cdot \Delta w) $

What is best way to find out $\Delta w$ or $\Delta b$ rather than making any guess?



- One of the way is to calculte the $\Delta \theta$ , by taking derivative of loss function w.r.t w & b.

# Algo:

- Initialize: $w $ & $b$
- Iterate over data: 
    - compute $\hat y$
    - compute Loss $\mathcal{L} (w,b)$
    - $w = w + \eta  \cdot \frac{\partial \mathcal{L}}{\partial w} $ \
    - $b = b + \eta \cdot \frac{\partial \mathcal{L}}{\partial b} $ \
    - till satisfied 
  
- Smaller steps taken from w, reduced the loss.
    $\mathcal{L} (w) > \mathcal{L} ( w + \eta \cdot \Delta w) $ \
    We need away which can give relate the function value before and after small steps.

    such relation we can get from taylor series,

    $f(x+\Delta x) = f(x) + \Delta x f'(x) + \frac{1}{2!}\Delta x^2 f''(x) + ...$


- Actually, loss function is function of $\mathcal{L}(w,b ) = \mathcal{L}(\theta)$. 
- Now writing the taylor series for loss function, \
    $\mathcal{L}(\theta + \eta u) = \mathcal{L}(\theta) +\eta \cdot u^T  \nabla_{\theta}\mathcal{L}(\theta) + \frac{\eta^2}{2!} \cdot u^T  \nabla_{\theta}^2 \mathcal{L}(\theta) u + ...$ \
    
    where, $\Delta \theta = u = \begin{bmatrix} \Delta w \\ \Delta b \end{bmatrix} $ , $ \nabla_{\theta} L = \begin{bmatrix} \frac{\partial L}{\partial w} \\ \frac{\partial L}{\partial b} \end{bmatrix}$

  - Ignoring the higher order terms, loss function can be written as, \
    $\mathcal{L}(\theta + \eta u) = \mathcal{L}(\theta) +\eta \cdot u^T  \nabla_{\theta}\mathcal{L}(\theta) $ \
    $\mathcal{L}(\theta + \eta u) - \mathcal{L}(\theta)  = \eta \cdot u^T  \nabla_{\theta}\mathcal{L}(\theta) $   ----$\diamond$
  - To achieve the goal $\mathcal{L} (w,b) > \mathcal{L} ( w + \eta \cdot \Delta w, b + \eta \cdot \Delta b) $\
   $\mathcal{L} (w,b) - \mathcal{L} ( w + \eta \cdot \Delta w, b + \eta \cdot \Delta b) > 0$ \
   $\mathcal{L} ( w + \eta \cdot \Delta w, b + \eta \cdot \Delta b) - \mathcal{L} (w,b) < 0$     ----$\star$ \
  - By comparing $\star$ and $\diamond$ it is clear that, \
  $\eta \cdot u^T  \nabla_{\theta}\mathcal{L}(\theta) < 0$  has to be negative. \
  $u^T  \nabla_{\theta}\mathcal{L}(\theta) < 0$     where $u^T,  \nabla_{\theta}\mathcal{L}(\theta)\in \mathcal{R^2}$

   $u^T  \nabla_{\theta}\mathcal{L}(\theta) < 0$

   The angle between two vector can be writtena as,
   $-1 \le \cos \beta = \frac{u^T  \nabla_{\theta}\mathcal{L}(\theta)}{||u || ||\nabla_{\theta}\mathcal{L}(\theta)||} \le 1 $  

   Now, multiplying both side with constant denominator $k = ||u || ||\nabla_{\theta}\mathcal{L}(\theta)||$,

   $-k \le u^T  \nabla_{\theta}\mathcal{L}(\theta)\le k $ 

   Now, goal is term $u^T  \nabla_{\theta}\mathcal{L}(\theta) < 0$

   The most negative value can be -1 which indicates $\beta = 180 $

   Which gives **gradient descent rule**.




# Gradinet Descent Rule

- The direction in which intend to move should be 180 w.r.t gradient
- or Move in opposite direction to gradient 



# Parameter update rule (Observe the negative sign ):

- $w_{t+1} = w_{t} - \eta \cdot \Delta w_t $ \
- $b_{t+1} = b_t - \eta \cdot \Delta b_t $ \

where, $\Delta w_t = \frac{\partial \mathcal{L}(w,b)}{\partial w}_{w=w_t,b=b_t}$ and $\Delta b_t = \frac{\partial \mathcal{L}(w,b)}{\partial b}_{w=w_t,b=b_t}$


**Why do we have negative sign?**
- because it satisifes the $\mathcal{L} ( \theta + u) - \mathcal{L} (\theta) < 0$ and we need k*$\cos \beta < 0 $ to be the most negative which happens when $\beta=180$. so, we need to move in the negative direction of the gradient to minimize the lost.


# Computational way to calculate gradients:

Calculation of gradient is necessary in loss function.

Loss function $\mathcal{L}$ = $\frac{1}{n}\sum_i^n (f(x_i) - y_i)^2$

$\frac{\partial \mathcal{L}}{\partial w} = \frac{\partial }{\partial w} \bigg[ \frac{1}{n}\sum_i^n (f(x_i) - y_i)^2\bigg]$

$\frac{\partial \mathcal{L}}{\partial w} =   \frac{1}{n}\sum_i^n \frac{\partial }{\partial w} (f(x_i) - y_i)^2$

Now, let's consider diffrential part,

$\frac{\partial \mathcal{L}}{\partial w} =   \frac{\partial }{\partial w} (\frac{1}{2}f(x) - y)^2$

$ = \frac{1}{2} \cdot 2 \cdot  (f(x) - y) \cdot \frac{\partial }{\partial w} (f(x) - y)$

$ =  (f(x) - y) \cdot \frac{\partial }{\partial w} (f(x) - y)$

$ =  (f(x) - y) \cdot \frac{\partial }{\partial w} f(x) - \frac{\partial }{\partial w} y$


$ =  (f(x) - y) \cdot \frac{\partial }{\partial w} f(x) $

$ =  (f(x) - y) \cdot \frac{\partial }{\partial w} f(x) $

$ =  (f(x) - y) \cdot \frac{\partial }{\partial w} \frac{1}{1+\exp^{-(wx+b)}} ------\star$

Solving derivative part,

$\frac{1}{1+\exp^{-(wx+b)}}$

$= \frac{1}{(1+\exp^{-(wx+b)})^2} \cdot \frac{\partial }{\partial w} {\exp^{-(wx+b)}}$

$= \frac{1}{(1+\exp^{-(wx+b)})^2} \cdot {\exp^{-(wx+b)}} \cdot \frac{\partial }{\partial w} -(wx+b)$

$= \frac{1}{(1+\exp^{-(wx+b)})^2} \cdot {\exp^{-(wx+b)}} \cdot (-x)$

$= \frac{1}{(1+\exp^{-(wx+b)})} \cdot \frac{(1+\exp^{-(wx+b)})}{(1+\exp^{-(wx+b)})} \cdot {\exp^{-(wx+b)}} \cdot (x)$

$= f(x) \cdot (1-f(x)) \cdot x$

Putting result in ($\star$) [It represent only for one term],

$\frac{\partial \mathcal{L}}{\partial w} = (f(x) - y) \cdot f(x) \cdot (1-f(x)) \cdot x$

for all term,

$\Delta w =  \frac{\partial \mathcal{L}}{\partial w} = \sum_i^n (f(x_i) - y_i) \cdot f(x_i) \cdot (1-f(x_i)) \cdot x_i$

**likewise for b**,

$\Delta b =  \frac{\partial \mathcal{L}}{\partial w} = \sum_i^n (f(x_i) - y_i) \cdot f(x_i) \cdot (1-f(x_i)) $
