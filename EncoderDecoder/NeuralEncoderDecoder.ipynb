**Language Modelling :**
- How sentences are constructed using models
- Prediction using history or prediction of $t^{th}$ word using $t-i$ words
- Given $y_1,y_2,y_3,... \rightarrow y* = argmax \Pr(y_t |y_1,y_2,...,y_{t-1})$

RNN.png

$\Pr(y_t = j | y_1^{t-1} = P(y_t = j |s_t) = softmax(Vs_t+c)_j$

# Example:
- India, officially the Republic of India, is a country in South Asia.
  - Data : All sentences from any large corpus
  - Model :
    - $ s_t = \sigma(W  s_{t-1} + U x_t + b)$
    - $P(y_t = j|y_1^{t-1} =softmax(Vs_t +c)_j$

  - Parameter : U,V,W,b,c

  - Loss :
    - $\mathcal L(\theta) = \sum_{t-1}^T \mathcal{L}(\theta)$
    - $\mathcal L(\theta) = -\log P(y_t = l_t|y_1^{t-1})$ --- $l_t$ is true class label

# Famous ShortHand:

RNN : 
  - $s_t = \sigma(Ux_t + Ws_{t-1}+b$
  - **ShortHand:** $s_t = RNN(s_{t-1},x_t)$

GRU :
  - $\tilde s_t = \sigma(W(o_t \odot s_{t-1}) + Ux_t + b)$
  - $s_t = i_t \odot s_{t-1} + (1-i_t) \odot \tilde s_t$
  - **ShortHand:** $s_t = GRU(s_{t-1},x_t)$

LSTM :
  - $\tilde s_t = \sigma(W h_{t-1}) + Ux_t + b)$
  - $s_t = f_t \odot s_{t-1} + i_t \odot \tilde s_t$
  - $h_t = o_t \odot \sigma (s_t)$
  - **ShortHand:** $h_t, s_t = LSTM(h_{t-1},s_{t-1},x_t)$

# Image Captioning:

- Task :  Given Image caption the image 
- data :
  - image -> Man throwing disc in the garden 
  - $\{x_i = image ,y_i = caption_i\}_{i=1}^N$
  - MSCOCO dataset
- Model:
  - $y = f(x)$
  - Encoder :
    - $s_0 = CNN(x_i)$
  - Decoder :

    - $s_t = RNN(s_{t-1},e(\hat y_{t-1})$

    - $P(y_t|y_1^{t-1,I}) = softmax(V s_t+b)$




model.jpg

    - CNN is used to encode the image 
    - RNN is then used to decode the image from encoding 
    - FNN is used to classify the words 

- Parameters : $U_{decoder}, V, W_{decoder}, W_{convoution}, b$

- Loss : 
  - $ \sum_{i=1}^T \mathcal L_t (\theta) = - \sum_{i=1}^T \log P(y_t = l_t | y_1{t-1},I)$
- Algorithm : Gradient Descent with backpropogation

# Machine Translation:

- Task :  Machine Translation 
- data :
  - $\{x_i = image ,y_i = caption_i\}_{i=1}^N$
- Model:
  - $y = f(x)$
  - Encoder :
    - $h_t = LSTM(h_{t-1},x_{it})$
  - Decoder :
    - $s_0 = h_T (T $ is lenght of input)

    - $s_t = LSTM(s_{t-1},e(\hat y_{t-1})$ (Model -1)
    - $s_t = LSTM(s_{t-1},e(\hat y_{t-1})$ (Model -2)

    - $P(y_t|y_1^{t-1,x}) = softmax(V s_t+b)$

- Parameters : $U_{decoder}, V, W_{decoder}, U_{encoder}, W_{encoder}, b$

- Loss : 
  - $ \sum_{i=1}^T \mathcal L_t (\theta) = - \sum_{i=1}^T \log P(y_t = l_t | y_1{t-1},I)$
- Algorithm : Gradient Descent with backpropogation
