# Object detection
- Find object
- Generate bounding box around the object


### Pipeline for object detection 
1. Identify the potencial region coordinates which contains the object
2. Generate mini-Images
3. Extract the mini-images from the features (SIFT,HOG,CNNs)
4. Pass images through image classifier to determine the class
5. Correct the proposed bounding box using regression problem
- Region predicted a',b',w',h'
- correct is a,b,w,h
- Find regression loss $\sum(y_i - \hat y)^2$
- Repeat till satisfied

$\min \sum_{i=1}^N \bigg( \frac{y* - y}{w} - w_2^Tz\bigg)^2$
  - $\frac{x*-x}{w}$ is the normalized difference between proposed x and x*
  - model predicts $w^Tz$ as the difference
  - The objective of the function is to minimize the differnece between the predicted and the actual value.
